```
# vamos carregar o arquivo com o resultado da eleição no Brasil dividido por município/zona
df_votacao = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Curso LABHacker/Dados/votacao_candidato_munzona_2018_BR.csv', sep=';', encoding='latin1')

df_votacao.head()
```

```
# visualizando o dataframe trasnsposto
df_votacao.head().T
```

```
# vamos separar os dados pois precisamos apenas dos dados do segundo turno
df_votacao = df_votacao[df_votacao['NR_TURNO'] == 2]

df_votacao.sample(3).T
```

```
# vamos verificar quantos municípios únicos existem no conjunto de dados
# vamos epgar por código do município
df_votacao['CD_MUNICIPIO'].nunique()
```

```
# o Brasil tem 5570 municípios
# vamos verificar algumas cidades fora do brasil
df_votacao[df_votacao['SG_UF'] == 'ZZ'].sample(4).T
```

```
# quantas cidades fora do Brasil estão no conjunto de dados
df_votacao[df_votacao['SG_UF'] == 'ZZ']['CD_MUNICIPIO'].nunique()
```

```
# vamos eliminar os municipios fora do Brasil
df_votacao = df_votacao[df_votacao['SG_UF'] != 'ZZ']
```

```
# verificando a quantidade de municipios
df_votacao['CD_MUNICIPIO'].nunique()
```

```
# os dados estão separados por município/zona
# municípios pequenos normalmente tem apenas 1 zona eleitoral
# mas municípios maiores podem ter várias zonas eleitorais

# vamos verificar Belo Horizonte
df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE'].T
```

```
# eliminar algumas colunas do dataframe que não serão utilizadas
df_votacao = df_votacao[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']]

df_votacao.head()
```

```
# precisamos agregar os dados por município e candidato
df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum()
```

```
# fazer o groupyby, reindexar o indice, salvar em um dataframe
df_votacao = df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum().reset_index()

df_votacao.sample(4)
```

```
# vamos verificar novamente Belo Horizonte
df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE']
```

```
# ordernar o dataframe pela quantidade de votos e depois deixar apenas o primeiro registro de cada município
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False).head(10)
```

```
# vamos ordenar
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)

# dropar as linhas duplicadas, matendo a primeira
df_votacao.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)

df_votacao.head()
```

```
# vamos verificar novamente Belo Horizonte
df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE']
```

```
# agora vamos criar uma coluna representando o resultado da eleição naquele município
# sendo que o resultado é True se o candidato Bolsonaro venceu naquele município
# e False em caso contrário
df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```
# verificando o resultado
df_votacao.head(10)
```

```
# mostrando alguns valores do dataframe do geopandas
df_mapa.sample(3)
```

```
# tentando encontrar o município pelo código
df_votacao[df_votacao['CD_MUNICIPIO'] == 3509304]
```

```
# porque não podemos usar o nome do município para juntar os dados
df_votacao[df_votacao['NM_MUNICIPIO'] == 'OURO BRANCO']
```

```
# recuperar a tabela de correpondência entre IBGE e TSE
# direto do github do estadão
link = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'

df_equivalencia = pd.read_csv(link)

df_equivalencia.head()
```

```
# vamos preparar os dataframes
```

```
# vamos reindexar o dataframe de votacao pelo código do TSE
df_vot_novo = df_votacao.copy()

df_vot_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True)

df_vot_novo.head()
```

```
# vamos reindexar o dataframe de equivalencia pelo código do TSe
df_equi_novo = df_equivalencia.copy()

df_equi_novo.set_index('COD_TSE', drop=False, inplace=True)

df_equi_novo.head()
```

```
# verificando o tamanho dos dataframes
df_vot_novo.info()
```

```
df_equi_novo.info()
```

```
# criando um novo dataframe com base no merge de votação e equivalência
df_vot_equi = df_vot_novo.join(df_equi_novo)

df_vot_equi.head()
```

```
# vamos juntar o dataframe do geopandas com o dataframe votação/equivalência
```

```
df_vot_equi.info()
```

```
df_mapa.info()
```

```
# transformar a coluna de código do ibge do dataframe de votação/equivalência
df_vot_equi['GEOCOD_IBGE'] = df_vot_equi['GEOCOD_IBGE'].astype(str)

df_vot_equi.info()
```

```
# vamos unificar os indices

# mudando o indice do dataframe do mapa
df_mapa.set_index('CD_GEOCMU', drop=False, inplace=True)
```

```
# mudando o indice do dataframe de votação/equivalência
df_vot_equi.set_index('GEOCOD_IBGE', drop=False, inplace=True)
```

```
# vamos unificar o dataframe do mapa com o dataframe de votação/equivalência
df_mapa_novo = df_mapa.join(df_vot_equi)

df_mapa_novo.head().T
```

```
# verificando o tamanho do mapa novo
df_mapa_novo.info()
```

```
# vamos descobrir os 2 municípios a mais no mapa
df_mapa_novo[df_mapa_novo['NR_CANDIDATO'].isnull()]
```

```
# matar a coluna ajuste
df_mapa_novo.drop(columns=['AJUSTE'], inplace=True)
```

```
# vamos eliminar a informação das 2 lagoas do RS
df_mapa_novo.dropna(inplace=True)
```

```
df_mapa_novo.info()
```

```
# vamos desenhar o mapa

# criar a referência para o código de cor (False/0 = red, True/1 = green)
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['red', 'green'])

# desenhando o mapa
fig, ax = plt.subplots(1, figsize=(12,12))

ax = df_mapa_novo.plot(column='RESULTADO', cmap=cmap, legend=True, linewidth=0.1, ax=ax, edgecolor='grey')

# escondendo os eixos
ax.set_axis_off()

# titulo
ax.set_title('Resultado Eleição 2018 por Município / 2º Turno', fontdict={'fontsize': '25', 'fontweight': '3'})

# rodapé
ax.annotate('Fonte: IBGE, 2015; TSE, 2019; Jornal o Estado de São Paulo, 2019', xy=(0.1, 0.08), 
            xycoords='figure fraction', horizontalalignment='left', verticalalignment='top', fontsize=10, color='#555555')

# altera legenda
leg = ax.get_legend()
leg.get_texts()[0].set_text('Fernando Haddad')
leg.get_texts()[1].set_text('Jair Bolsonaro')

plt.show()
```
