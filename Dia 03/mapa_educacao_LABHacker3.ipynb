# Mapa das escolas do RS por taxa de distorção de série

```
# importando as bibliotecas
import pandas as pd
%matplotlib inline
import folium
```

```
# carregar o arquivo csv
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Curso LABHacker/Dados/fee-2013-mun-taxa-de-distorcao-idade-serie-total-102524.csv', encoding='latin1', skiprows=1)

df.head()
```

```
# renomeando colunas
df.rename(columns={'/Educação/Ens...de Série/Total 2013 (-)': 'tx_distorcao'}, inplace=True)

df.head()
```

```
# verificando os dataframe
df.info()
```

```
# converter tx_distorcao para float
# primeiro vamos trocar ',' por '.'
df['tx_distorcao'] = df['tx_distorcao'].str.replace(',', '.')

df.head()
```

```
# fazendo a conversão
df['tx_distorcao'] = df['tx_distorcao'].astype(float)

df.info()
```

```
# 10 municipios com menor taxa de distorção
df.nsmallest(10, 'tx_distorcao')
```

```
# 10 municipios com maior taxa de distorção
df.nlargest(10, 'tx_distorcao')
```

```
# plotar o histograma da taxa de distorção
df['tx_distorcao'].plot.hist(bins=100)
```

```
# quantidade de municipios com taxa de distoção menor ou igual a 10
df[df['tx_distorcao'] <= 10].count()
```

```
# quantidade de municipios com taxa de distoção maior ou igual a 45
df[df['tx_distorcao'] >= 45].count()
```

```
# vamos aprender como o folium funciona

# criar um gráfico, com base em uma coordenada central
brasil = folium.Map(
    location=[-13.6561589, -69.7309264], # coordenadas do google maps
    zoom_start=4
)

# mostrando o mapa
brasil
```

```
# vamos criar o mapa do RS
rs = folium.Map(
    location=[-30.5, -52], # coordenadas ajustadas do google maps
    zoom_start=7
)

# mostrando o mapa do rs
rs
```

```
# percorrer o dataframe com base nas melhores escolas (tx_distorcao <=10)
# e desenhar um marcador para cada escola
# usando for e iterrows
for indice, municipio in df[df['tx_distorcao'] <= 10].iterrows():
  folium.Marker(
      location=[municipio['latitude'], municipio['longitude']],
      popup=municipio['Município'],
      icon=folium.map.Icon(color='green')
  ).add_to(rs)

rs
```

```
# percorrer o dataframe com base nas melhores escolas (tx_distorcao >= 45)
# e desenhar um marcador para cada escola
# usando for e iterrows
for indice, municipio in df[df['tx_distorcao'] >= 45].iterrows():
  folium.Marker(
      location=[municipio['latitude'], municipio['longitude']],
      popup=municipio['Município'],
      icon=folium.map.Icon(color='red')
  ).add_to(rs)

rs
```
