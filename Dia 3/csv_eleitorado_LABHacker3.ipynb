# Análise do Eleitorado Brasileiro

```
# importando as bibliotecas
import pandas as pd
```

```
# carregar o arquivo csv
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Curso LABHacker/Dados/eleitorado_municipio_2020.csv', sep=';', encoding='latin1')

df.head(3)
```

```
# tipos e quantidades
df.info()
```

```
# as 5 cidades com maior quantidade de eleitores deficientes
df.nlargest(5, 'QTD_ELEITORES_DEFICIENTE').T
```

```
print('Eleitoras: ', df['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores: ', df['QTD_ELEITORES_MASCULINO'].sum())
print('Naõ informado: ', df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```
# criar variaveis para calcular o percentual
tot_eleitores = df['QTD_ELEITORES'].sum()
tot_fem = df['QTD_ELEITORES_FEMININO'].sum()
tot_masc = df['QTD_ELEITORES_MASCULINO'].sum()
tot_nao = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```
# mostrar os valores percentualmente
print('Eleitoras: ', ((tot_fem/tot_eleitores)*100).round(2))
print('Eleitores: ', ((tot_masc/tot_eleitores)*100).round(2))
print('Naõ informado: ', ((tot_nao/tot_eleitores)*100).round(2))
```

```
# quantos municipios com amsi homens que mulheres
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```
# criar uma nova coluna para indicar a relação fem/masc
df['RELACAO_FM'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```
# quais são os municipios com maior relação f/m
df.nlargest(5, 'RELACAO_FM').T
```

```
# quais os municipios com a menor relação f/m
df.nsmallest(5, 'RELACAO_FM').T
```

```
# criar um novo DataFrame só com municipios do Brasil
df_brasil = df[df['CD_PAIS'] == 1].copy()

df_brasil.head(3).T
```

```
# quais são os municipios com maior relação f/m
df_brasil.nlargest(5, 'RELACAO_FM').T
```

```
# quais os municipios com a menor relação f/m
df_brasil.nsmallest(5, 'RELACAO_FM').T
```

```
%matplotlib inline

# plotar um gráfico de distribuição da relação f/m
# bins = quantidade de barras
df_brasil['RELACAO_FM'].plot.hist(bins=100)
```

```
# importando as bibliotecas gráfias
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# plotar a distribuição da relação f/m
# kde=False -> ele tirar a curva de aproximação
sns.distplot(df_brasil['RELACAO_FM'], bins=100, color='red', kde=False)

# embelezando o gráfico
plt.title('Relação Eleitoras/Eleitores', fontsize=18)
plt.xlabel('Eleitoras/Eleitores', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.axvline(1.0, color='black', linestyle='--')
```

```
# verificar a distribução f/m mas mostrando os pontos (municipios)
sns.swarmplot(data=df_brasil, x='NM_REGIAO', y='RELACAO_FM')
plt.axhline(1.0, color='black', linestyle='-')
```
