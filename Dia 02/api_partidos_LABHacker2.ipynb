# Análise dos dados da API de partidos
Análise dos dados obtidos pela API da Câmara dos Deputados

Link -> https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla


```
# importando as bibliotecas
import pandas as pd
import requests
```

```
# obter os dados da API da câmara dos deputados
# variavel para url
url = 'https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla'

# obtendo os dados
resposta = requests.get(url)

# verificando a resposta
resposta.text
```

```
# converter a resposta da API para DataFrame
df = pd.DataFrame(resposta.json()['dados'])

df.head()
```

```
# qual o numero atual de partidos politicos
df['sigla'].count()
```

```
df['nome']
```

```
# qualquer campo do DataFrame pode ser obtido usando as funções de localização
# o iloc retorna um campo com base no seu índice [linha, coluna]

df.iloc[0, 3]
```

```
# o loc retorna o campo com base no nome da coluna
df.loc[0, 'uri']
```
